## Extract Stock & Index codes, Top gainers, Top losers from NSE

```python
from nsetools import Nse
```

```python
nse=Nse()
print(nse)
```

## Getting a stock code

```python
import json
q = nse.get_quote('infy')
```

```python
display(q)
```

## Get Index List from NSE

```python
nse.get_index_list()
```

```python
nse.get_index_quote("nifty bank")
```

```python
nse.get_index_quote("nifty 50")
```

## Get list of all traded stocks from NSE

```python
nse.get_stock_codes()
```

## Advance Decline

```python
adv_dec = nse.get_advances_declines()
display(adv_dec)
```

## Top Gainers

```python
top_gainers = nse.get_top_gainers()
print(json.dumps(top_gainers,indent=4))
```

## Top losers

```python
top_losers = nse.get_top_losers()
print(json.dumps(top_losers,indent=4))
```
